```python
import numpy as np 
import matplotlib.pyplot as plt
```

### ONLY ALIGNED

```python
#20x20 lattice 
#plotting average energy per spin as a function of MC cycles for two temperatures: T = 1 and T = 2.4

#import quantities 
#need to import data for L=20, mc1000000, burn0 and t1
#need to import data for L=20, mc1000000, burn0 and t2.4

data_t1_align = np.loadtxt("./2/qt_L2_A1_mc1000000_burn0_t1.000000.txt")
data_t24_align = np.loadtxt("./2/qt_L2_A1_mc1000000_burn0_t1.000000.txt")

data_t1_random = np.loadtxt("./2/qt_L2_A0_mc1000000_burn0_t1.000000.txt")
data_t24_random = np.loadtxt("./2/qt_L2_A0_mc1000000_burn0_t1.000000.txt")
```

```python
#import quantities for T = 1, aligned
e_1_align = data_t1_align[:,0] #average energy per spin 
m_1_align = data_t1_align[:,1] #average magnetisation per spin 

#import quantities for T=2.4, aligned
e_24_align = data_t24_align[:,0] #average energy per spin 
m_24_align = data_t24_align[:,1] #average magnetisation per spin 

#import quantities for T = 1, random
e_1_random = data_t1_random[:,0] #average energy per spin
m_1_random = data_t1_random[:,1] #average magnetisation per spin

#import quantities for T = 2.4, random
e_24_random = data_t24_random[:,0] #average energy per spin
m_24_random = data_t24_random[:,1] #average magnetisation per spin
```

```python
#plotting for T = 1, aligned and random 
# extract every 50th value for plotting
e_1_align = e_1_align[::50]
e_1_random = e_1_random[::50]
plt.plot([(i+1)*50 for i in range(len(e_1_align))], e_1_align, label = "aligned")
plt.plot([(i+1)*50 for i in range(len(e_1_random))], e_1_random, label = "random")
plt.xscale("log")
#plot with grid
plt.grid()
plt.xlabel("MC cycles")
plt.ylabel("$\langle \epsilon \\rangle \;[J]$")
```
